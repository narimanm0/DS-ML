```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv(r'Almond.csv')
data
```

```python
for i in data.drop(["Type"], axis=1):
    data[i] = data[i].fillna(data[i].mean())
```

```python
data.isnull().sum()
```

```python
data["Type"].unique()
```

```python
data = data.drop(["Unnamed: 0"], axis=1)
```

```python
data["Type"] = data["Type"].map({
    "REGULAR": 0,
    "SANORA": 1,
    "MAMRA": 2
})
```

```python
data.corr()["Type"]
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

variables = data[[
    # 'Length (major axis)',
    # 'Width (minor axis)',
       # 'Thickness (depth)',
    'Area',
    # 'Perimeter',
    'Roundness',
    # 'Solidity',
       'Compactness',
    # 'Aspect Ratio',
    # 'Eccentricity',
    # 'Extent',
       # 'Convex hull(convex area)'
]]

vif = pd.DataFrame()
vif["vif"] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]
vif["features"] = variables.columns
vif
```

```python
data = data[[
    'Area',
    'Roundness',
    'Compactness',
    "Type"
]]
```

```python
for i in [
    'Area',
    'Roundness',
    'Compactness'
]:
    sns.boxplot(data=data, x=data[i])
    plt.show()
```

```python
q1 = data.quantile(0.25)
q3 = data.quantile(0.75)
IQR = q3-q1
Upper = q3+1.5*IQR
Lower = q1-1.5*IQR
```

```python
for i in [
    'Area',
    'Roundness',
    'Compactness'
]:
    data[i] = np.where(data[i]>Upper[i], Upper[i],data[i])
    data[i] = np.where(data[i]<Lower[i], Lower[i], data[i])
```

```python
for i in [
    'Area',
    'Roundness',
    'Compactness'
]:
    sns.boxplot(data=data, x=data[i])
    plt.show()
```

```python
y = data["Type"]
x = data.drop(["Type"], axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=47)
```

```python
from sklearn import metrics
from sklearn.metrics import roc_auc_score

from sklearn.metrics import roc_auc_score

from sklearn.metrics import roc_auc_score

def evaluate(model, x_test, y_test):
    y_pred = model.predict(x_test)   # Predicted class labels
    y_prob = model.predict_proba(x_test)  # Probabilities for all classes

    # Ensure y_test is in the correct format
    if len(y_test.shape) == 2 and y_test.shape[1] > 1:
        y_test = y_test.argmax(axis=1)  # Convert one-hot encoded to class labels

    # Compute ROC AUC score for multiclass classification
    roc_prob = roc_auc_score(y_test, y_prob, multi_class='ovr')  
    
    gini_prob = roc_prob * 2 - 1

    print('Model Performance')
    print(f'Gini Coefficient: {gini_prob:.4f}')

```

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
log_model.fit(x_train, y_train)
```

```python
evaluate(log_model, x_test, y_test)
```
