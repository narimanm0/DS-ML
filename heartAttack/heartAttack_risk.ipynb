```python
import pandas as pd
from pandas import set_option
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Reading data and setting option for seeing all columns we have...

```python
data = pd.read_csv(r'heart_attack_dataset.csv')
pd.set_option('display.max_columns', 30)
```

```python
data
```

```python
data.describe(include="all")
```

We have two columns that have numerous unique values: "Patient ID" and "Blood Pressure"

```python
data = data.drop(["Patient ID"], axis=1)
```

We have x/y in "Blood Pressure". We create two new columns to demonstrate x and y seperately and integer form.

```python
data[["systolic_pressure", "diastolic_pressure"]] = data["Blood Pressure"].str.split('/', expand=True)
```

```python
data["systolic_pressure"] = data["systolic_pressure"].astype(int)
data["diastolic_pressure"] = data["diastolic_pressure"].astype(int)
```

```python
data = data.drop(["Blood Pressure"], axis=1)
```

Let's remove "Country" column also as we have "Continent" columns with a little amount of unique values.

```python
data = data.drop(["Country"], axis=1)
```

Checking whether we have NaN values or not

```python
data.isnull().sum()
```

Creating a list of column names which are categorical

```python
data_cat = data.select_dtypes(include='object').columns.tolist()
```

```python
data_cat
```

```python
data_num = data.drop(data_cat, axis=1)
```

Checking CORRELATION

```python
data_num.corr()["Heart Attack Risk"]
```

```python
data = data.drop("Family History|Previous Heart Problems|BMI".split("|"), axis=1)
```

```python
data_num = data_num.drop("Family History|Previous Heart Problems|BMI".split("|"), axis=1)
```

checking VIF

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

variables = data_num[[
    'Age',
    # 'Cholesterol',
    # 'Heart Rate',
    'Diabetes',
    # 'Smoking',
    'Obesity',
       'Alcohol Consumption',
    'Exercise Hours Per Week',
    'Medication Use',
       'Stress Level',
    'Sedentary Hours Per Day',
    'Income', 'Triglycerides',
       'Physical Activity Days Per Week',
    # 'Sleep Hours Per Day',
    # 'systolic_pressure',
    # 'diastolic_pressure'
]]

vif = pd.DataFrame()
vif["VIF"] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]
vif["Features"] = variables.columns
vif
```

```python
data = data.drop([
    'Cholesterol',
    'Heart Rate',
    'Smoking',
    'Sleep Hours Per Day',
    'systolic_pressure',
    'diastolic_pressure'
], axis=1)
```

```python
data_num = data_num.drop([
    'Cholesterol',
    'Heart Rate',
    'Smoking',
    'Sleep Hours Per Day',
    'systolic_pressure',
    'diastolic_pressure'
], axis=1)
```
