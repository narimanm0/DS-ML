```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data = pd.read_excel(r'paris_house_prices.xlsx')
data
```

```python
data.describe(include="all")
```

```python
data.isnull().sum()
```

```python
data_cat = data.select_dtypes(include="object").columns.tolist()
```

```python
data_num = data.drop(data_cat, axis=1)
```

```python
data_num.corr()["price"]
```

```python
data = data.drop(['numberOfRooms', 'floors', 'cityCode', 'cityPartRange',
       'made', 'basement', 'attic', 'hasGuestRoom'], axis=1)
```

```python
data_num = data_num.drop(['numberOfRooms', 'floors', 'cityCode', 'cityPartRange',
       'made', 'basement', 'attic', 'hasGuestRoom'], axis=1)
```

```python
data_num.columns
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

variables = data[[
    'squareMeters', 
    'numPrevOwners',
    'garage',
    'price'
]]

vif = pd.DataFrame()
vif["VIF"] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]
vif["Feature"] = variables.columns
vif
```

```python
for i in data_num:
    sns.boxplot(data=data, x=data[i])
    plt.show()
```
