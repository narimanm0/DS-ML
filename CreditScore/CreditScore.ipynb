```python
import pandas as pd
from pandas import set_option
pd.set_option('display.max_columns',90)

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv(r'credit_score.csv')
data
```

```python
data.describe(include="all")
```

```python
data = data.drop("ID CustomerID SSN TypeofLoan".split(), axis=1)
```

```python
data
```

```python
data["CreditScore"] = np.where(data["CreditScore"]=="Poor", 1, 0)
```

```python
data.describe(include="all")
```

```python
data_cat = data.select_dtypes(include="object").columns.tolist()
```

```python
data.isnull().sum()
```

```python
data_num = data.drop(data_cat, axis=1)
```

```python
for i in data_cat:
    data[i] = data[i].fillna(data[i].mode()[0])
```

```python
for i in data_num.columns:
    data[i] = data[i].fillna(data[i].mean())
```

```python
data.isnull().sum()
```

```python
data_num.corr()["CreditScore"]
```

```python
corr_deleted = "Age AnnualIncome NumBankAccounts NumCreditCard InterestRate NumofLoan NumofDelayedPayment NumCreditInquiries TotalEMIpermonth".split()
```

```python
data = data.drop(corr_deleted, axis=1)
```

```python
data_num = data_num.drop(corr_deleted, axis=1)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

variables = data[[
    'MonthlyInhandSalary',
    'Delayfromduedate',
    'ChangedCreditLimit',
       'OutstandingDebt',
    'CreditUtilizationRatio',
    'Amountinvestedmonthly',
       'MonthlyBalance',
    'CreditScore'
]]

vif = pd.DataFrame()
vif["VIF"] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]
vif["Features"] = variables.columns
vif
```
